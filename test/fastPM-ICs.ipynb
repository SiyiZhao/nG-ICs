```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = np.loadtxt("/home/siyizhao/SeeCosmology/data/pkl_z0/fastpm.dat")
data2 = np.loadtxt("/hscratch/siyizhao/ICs/fastpm/ics/test_pk_linear.txt")
data3 = np.loadtxt("/hscratch/siyizhao/ICs/fastpm/ics/test_pk_0.0200.txt")
k1, pk1 = data1[:, 0], data1[:, 1]
k2, pk2 = data2[:, 0], data2[:, 1]
k3, pk3 = data3[:, 0], data3[:, 1]
```

```python
boxL = 1000
Ngrid = 512
k_nyquist = np.pi / (boxL / Ngrid)
```

```python
plt.plot(k1, pk1, label=f'input: z=0')
plt.plot(k2, pk2, label=f'output: linear')
plt.plot(k3, pk3, label=f'output: z=49, a=0.02')
plt.plot(k1, pk1*0.02573235**2, label='g(z=49) x input')
plt.axvline(k_nyquist, 0,1, linestyle=':', label='k_Nyquist')
plt.xlabel(r'$k [h/{\rm Mpc}]$')
plt.xscale('log')
plt.xlim(0.005, 10)
plt.ylabel(r'$P(k) [({\rm Mpc}/h)^3]$')
plt.yscale('log')
plt.legend()
```
